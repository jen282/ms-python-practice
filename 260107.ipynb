```python
#복습1
#소수 판별 is_prime(), 입력값은 리스트[2,3.4,5.6,7,8,9]
def is_prime(num_list):
    ans = []
    for num in num_list:
        is_prime_num = True
        for i in range(2, num):
            if num%i==0:
                is_prime_num = False
                break
        ans.append(is_prime_num)
    return ans
```

```python
is_prime([2,3,4,5,6,7,8,9])
```

```python
#복습2
#등수 구하기
def rank(score):
    print(score)
    # avg = [(i+j)/2 for i,j in score]
    avg = [sum(s)/len(s) for s in score]
    print(avg)
    sorted_avg = sorted(avg, reverse=True) #sort()를 하면 원본을 변형, 원래 리스트 순서대로 랭크를 넣을 수가 없음
    # ranks = [sorted_avg.index(a) + 1 for a in avg] #enumerate를 사용하는 것이 훨씬 빠름
    ranks = [list(enumerate(sorted_avg))[0][0] for a in avg]
    print('등수: ', ranks)
    return ranks
```

```python
score = [[80, 70], [90, 50], [40, 70], [50, 80]]
avg = [(i+j)/2 for i,j in score]
print(avg)
sorted_avg = sorted(avg, reverse=True)
print(sorted_avg)
print(list(enumerate(sorted_avg))[0][0])
# ranks = [list(enumerate(sorted_avg))[a][0] for a in avg]
```

```python
print('score1')
score1 = [[80, 70], [90, 50], [40, 70], [50, 80]]
rank(score1) # [1, 2, 4, 3]

print('score2')
score2 = [[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]]
rank(score2) # [4, 4, 6, 2, 2, 1, 7]
```

## 클래스

* 클래스 = 붕어빵틀   객체 = 붕어빵
* 동작=메소드   속성=변수

```python
#클래스 이름 첫글자는 대문자! 다른 메서드, 변수와 구분하기 위함
class Person: 
    def __init__(self, name, age):  #self:이스턴스 간 구분을 위함
        self.name = name
        self.age = age
    def walk(self):
        print(f'{self.name}이(가) 걸어가요')
```

```python
person1 = Person('홍길동', 35)
```

```python
type(person1)
```

```python
print(type(person1)) #print를 찍어야 클래스라고 출력됨
```

```python
person2 = Person('김영희', 28)
```

```python
person2.walk()
```

### 상속

```python
class NewPerson(Person): # Person 클래스를 상속 
    # pass
    # 1. 기능확장
    def fly(self):
        print(f'{self.name}이 하늘을 날아요')
    # 2. 메서드 오버라이딩 : 같은 이름의 상속 받은 매서드를 재정의
    def walk(self):
        print(f'{self.name}이(가) 아주 빠르게 걸어가요')
    
```

```python
# 상속 받은 하위 클래스는 상위 클래스의 속성, 메서드를 상속 받아 사용할 수 있음 
person3 = NewPerson('김영광', 35)
person3.walk()
print(person3.age)
```

```python
person3.fly()
```

```python
class NewNewPerson(Person): # Person 클래스를 상속 
    # 4. 클래스 변수
    population = 0

    # 3. 오버로딩 : 
    def __init__(self, name, age, hobby): 
       super().__init__(name, age)  #부모 클래스에서 정의된 필드를 그대로 가져옴
       self.hobby = hobby
       # 4. 클래스 변수
       NewNewPerson.population += 1

    def fly(self):
        print(f'{self.name}이 하늘을 날아요')

    def walk(self):
        print(f'{self.name}이(가) 아주 빠르게 걸어가요')
        
    # 3. 오버로딩 
    def introduce(self):
        print(f'{self.name}의 취미는 {self.hobby}입니다')
```

```python
person4 = NewNewPerson('김남길', 45, "운동하기")
```

```python
person4.hobby
```

```python
person4.introduce()
```

```python
# 4. 클래스 변수 
person5 = NewNewPerson('김남길', 45, "운동하기")
person6 = NewNewPerson('수지', 32, "운동하기")
person7 = NewNewPerson('김태희', 45, "운동하기")
```

```python
#같은 클래스의 모든 객체가 같은 값을 가짐
print(person5.population)
print(person6.population)
print(person7.population)
```
