```python
# 2  
# Todo: 변수 a에 저장된 값의 b자리 값을 계산하여 변수 c에 저장하는 코드를 설계하라  
#예를 들어 a에 1234가 저장되어 있고 b가 1일 경우 첫 번째 자리값인 4가 c에 저장된다.  
#a가 1234일때 b가 4일 경우는 4번째 자리값인 1이 c에 저장되어야 한다
a = 1234 
b = 4
c = (a//10**(b-1))%10
print(c)
```

```python
# 3 
# 삼각형의 넓이 구하기 
# 사용자에게 삼각형의 밑변과 높이를 입력 받아  
# 삼각형의 넓이를 출력하는 코드를 작성하세요.
base = int(input('밑변의 길이를 입력하세요. '))
height = int(input('높이를 입력하세요. '))
area = base*height*0.5
print('넓이: ', area)
```

```python
# 4 
# 다음 그림과 같이 바둑돌의 개수를 늘려가며 삼각형 모양을 만들려고 합니다. 
# n이 입력으로 주어질 때, n단계에서 사용된 바둑돌의 개수를 출력하는 코드를 작성하세요.

n = int(input('단계를 입력하세요 '))
cnt = n*(n+1)*0.5
print('개수: ', cnt)
```

```python
# 5
# 정수 num1, num2가 입력으로 주어질 때, 첫 번째 줄에는 num1과 num2를 곱한 
# 수를, 두 번째 줄에는 num1을 num2번 반복하여 출력하도록 코드를 작성해주세요 
# 입출력 예 
# num1이 4, num2가 3이므로 첫째 줄에 4와 3을 곱한 12를 출력하고, 둘째 줄에 
# 4를 3번 반복한 "444"를 출력합니다.
num1 = int(input('첫번째 숫자를 입력하세요: '))
num2 = int(input('두번째 숫자를 입력하세요: '))
print(num1*num2)
for i in range(num2):
    print(num1, end='')
```

```python
# 성적처리 프로그램
score = int(input('학생의 점수를 입력 '))

if score > 90:
    grade = 'A'
elif 80 <= score < 90:
    grade = 'B'
elif 70 <= score < 80:
    grade = 'C'
elif 60 <= score < 70:
    grade = 'D'
elif score < 60:
    grade = 'E'
print(grade)
```

```python
# 9
# 구구단 전체(2단부터 9단까지)를 출력 
for i in range(2,10):
    print(f'{i}단')
    for j in range(1,10):
        print(f'{i} X {j} = {i*j}')
    print()
```

```python
# 10
# *을 가로로 a개, 세로로 b개 사용하여 직사각형을 만들려고 합니다. 정수 a, b가 
# 매개변수로 주어질 때, a x b 직사각형을 출력하도록 코드를 작성하세요 
a = int(input('가로 개수를 입력하세요'))
b = int(input('세로 개수를 입력하세요'))
for h in range(b):
    for w in range(a):
        print('*', end='')
    print()
```
