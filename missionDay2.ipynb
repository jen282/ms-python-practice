```python
#13
#딕셔너리 - 파이썬 용어 사전
dictionary = {'python':'a high level programming language',
              'loop': 'a sequence of instructions that are continually repeated'}
while True:
    print('파이썬 용어 사전')
    word = input('찾고 싶은 파이썬 용어를 입력하세요')
    if word in dictionary.keys():
        meaning = dictionary[word]
    elif word == '종료':
        break
    else:
        meaning = "사전에 찾는 단어가 없어요"
    print(f"입력단어->'{word}' is: {meaning}")
```

```python
#17
#가위바위보 게임 만들기
import random

sel = ['가위', '바위', '보']
def checkWin(u, c): #user가 이기면 참
    print(f'사용자: {u}, 컴퓨터: {c}')
    isUserWin = False
    if u=='가위':
        isUserWin=(c=='보')
    elif u=='바위':
        return c=='가위'
    elif u=='보':
        return c=='바위'

def continueConfirm():
    iscontinue = input('계속하시려면 y를 입력하세요')
    if iscontinue=='y' or iscontinue=='Y':
        return True
    else: return False
while True:
    user = input("가위, 바위, 보 : ")
    com = sel[random.randint(0,2)]
    if checkWin(user,com): #user가 이기면 참
        if continueConfirm():
            continue
        else:
            break
```
