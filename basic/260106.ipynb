```python
#복습
#20보다 작은 정수 주에서 3배수와 5배수를 제외하고 출력해 봅시다.
#단, 무한 반복 while true문 사용 
x=0

while True:
    x+=1
    if x==20:
        break
    elif x%3!=0 and x%5!=0:
        print(x, end=" ")
```

```python
#리스트 복습 미션
# 리스트에서 연속적으로 나타나는 숫자는 하나만 남기고 전부 제거
# 제거된 후 남은 수들은 기존 리스트의 순서를 유지해야 함
# numbers = [1,1,3,3,0.1,1] -> [1, 3, 0, 1]

numbers = [1,1,3,3,0,1,1]
result = [numbers[0]]
for num in numbers[1:]:
   if num!=result[-1]:
      result.append(num)
print(result)
```

```python
#도전미션
#전화번호부 미션
addr = {
    '박지민': '010-1111-1234', '전정국': '010-2222-1234',
    '민윤기': '010-3333-1234', '정호석': '010-4444-1234',
    '김태형': '010-5555-1234', '김석진': '010-6666-1234',
    }
while True:
    order = int(input('>메뉴 선택(1:추가 2:검색 0:종료)? '))
    if order == 0:
        print('0번 선택)')
        print('프로그램을 종료합니다!')
        break
    elif order ==1:
        print('1번 선택)')
        addName = input('친구 이름 입력>> ')
        print(f'친구 이름 입력>> {addName}')
        addPhone = input('전화번호 입력>> ')
        print(f'전화번호 입력>> {addPhone}')
        addr[addName]=addPhone
    elif order ==2:
        print('2번 선택)')
        print(f'검색가능한 친구들: {list(addr.keys())}') #리스트로 감싸면 리스트로 형변환
        friendName = input('친구이름 입력>> ')
        print(f'친구이름 입력>> {friendName}')
        #친구이름 존재 여부 확인 
        if friendName in addr:
            print(f'{friendName}의 연락처: {addr.pop(friendName)}')
        else:
            print(f'{friendName}의 연락처는 없습니다.')
    else:
        print('메뉴 선택 오류입니다. 다시 선택하세요.')
```

1. map(function, iterable)
* iterable의 아이템을 function에 인자로 전달
* function 파라미터 개수와 iterable 개수가 같아야 함
* 결과: map 타입
* 파이썬 자체가 메모리에 값을 직접 던지는 게 아니라 주소값만 저장하기 때문에 하나씩 순회하는 것은 시간이 걸릴 수 있어 map같은 함수를 제공해줌
그래서 map type을 출력하면 메모리값이 출력 -> list()로 변환해서 출력을 하는 것이 필요함

```python
myList = [1,2,3,4,5]

#1. for문 -> 속도 느림
result1=[]
for val in myList:
    result1.append(val+1)
print(f'result1 : {result1}')

#2. map함수 이용
result2 = []
def add_one(n):
    return n+1
result2 = map(add_one, myList)
print(f'result2 : {list(result2)}')
```

```python
# 실수를 가까운 정수값으로 변환하기
# map을 사용하지 않는다면 for문으로 하나하나 형변환을 해줘야 하겠지만 
# map을 이용하면 한 번에 변환 가능 
f = (3.14, -5.625, 100.4, 25.8)
# i = list(map(round, f))
# i
print(*map(round, f)) # * = unpacking 연산자, 객체를 list로 변환
```

```python
# 여러 기능을 선택해서 사용
def add_mul(mode, *args): #mode는 고정인자, args는 가변인자
    if mode=='add':
        result=0
        for i in args:
            result += i
            return result #return(단 함수 종료)이 여러개 있어도 괜찮음.
    elif mode=='mul':
        result=1
        for i in args:
            result*=i
    else:
        print('연산자 종류 확인 요망')
        result=None
    return result
```

```python
add_mul('add', 1,2)
```

```python
add_mul('mul', 1,2,3,4)
```

sorted vs sort
* sorted
    - 정렬된 결과를 반환
    - sorted_list = sorted(mylist)
* sort
    - 원본을 변형
    - mylist.sort()
    
-> 공통 파라미터
1. reverse = True / False
2. key = 
    - 람다 사용 가능 
    - 이 키값을 기준으로 정렬
    - 한 가지 예시만 키값에 넣어주면 그 기준에 맞춰 정렬 가능

```python
#sorted
my_list = [3, 1, 4, 1, 5, 9, 2]
sorted_list = sorted(my_list)  # 오름차순 정렬
print(sorted_list)
```

```python
#sort
my_list = [3, 1, 4, 1, 5, 9, 2]
my_list.sort()
my_list
```

```python
#key 값
str_list = ['좋은하루','good_morning','굿모닝','niceday']
print('기본 정렬', sorted(str_list))
str_list = ['좋은하루','good_morning','굿모닝','niceday']
print('길이로 정렬', sorted(str_list, key=len))
print('두번째 글자로 정렬', sorted(str_list, key=lambda x:x[1]))
#여러 개의 요소를 가진 경우
tuple_list = [('좋은하루', 0),
    	          ('niceday', 1), 
    	          ('좋은하루', 5), 
    	          ('good_morning', 3), 
    	          ('niceday',5)]
print('두 번째 요소 기준 정렬:\n'
      ,sorted(tuple_list, key=lambda x:x[1]))
```
