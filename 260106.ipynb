복습

```python
#복습
#20보다 작은 정수 주에서 3배수와 5배수를 제외하고 출력해 봅시다.
#단, 무한 반복 while true문 사용 
x=0

while True:
    x+=1
    if x==20:
        break
    elif x%3!=0 and x%5!=0:
        print(x, end=" ")
```

```python
#리스트 복습 미션
# 리스트에서 연속적으로 나타나는 숫자는 하나만 남기고 전부 제거
# 제거된 후 남은 수들은 기존 리스트의 순서를 유지해야 함
# numbers = [1,1,3,3,0.1,1] -> [1, 3, 0, 1]

numbers = [1,1,3,3,0,1,1]
result = [numbers[0]]
for num in numbers[1:]:
   if num!=result[-1]:
      result.append(num)
print(result)
```

```python
#도전미션
#전화번호부 미션
addr = {
    '박지민': '010-1111-1234', '전정국': '010-2222-1234',
    '민윤기': '010-3333-1234', '정호석': '010-4444-1234',
    '김태형': '010-5555-1234', '김석진': '010-6666-1234',
    }
while True:
    order = int(input('>메뉴 선택(1:추가 2:검색 0:종료)? '))
    if order == 0:
        print('0번 선택)')
        print('프로그램을 종료합니다!')
        break
    elif order ==1:
        print('1번 선택)')
        addName = input('친구 이름 입력>> ')
        print(f'친구 이름 입력>> {addName}')
        addPhone = input('전화번호 입력>> ')
        print(f'전화번호 입력>> {addPhone}')
        addr[addName]=addPhone
    elif order ==2:
        print('2번 선택)')
        print(f'검색가능한 친구들: {list(addr.keys())}') #리스트로 감싸면 리스트로 형변환
        friendName = input('친구이름 입력>> ')
        print(f'친구이름 입력>> {friendName}')
        #친구이름 존재 여부 확인 
        if friendName in addr:
            print(f'{friendName}의 연락처: {addr.pop(friendName)}')
        else:
            print(f'{friendName}의 연락처는 없습니다.')
    else:
        print('메뉴 선택 오류입니다. 다시 선택하세요.')
```

1. map(function, iterable)
* iterable의 아이템을 function에 인자로 전달
* function 파라미터 개수와 iterable 개수가 같아야 함
* 결과: map 타입

```python
myList = [1,2,3,4,5]

#1. for문 -> 속도 느림
result1=[]
for val in myList:
    result1.append(val+1)
print(f'result1 : {result1}')

#2. map함수 이용
result2 = []
def add_one(n):
    return n+1
result2 = map(add_one, myList)
print(f'result2 : {list(result2)}')
```

```python
# 실수를 가까운 정수값으로 변환하기
f = (3.14, -5.625, 100.4, 25.8)
i = list(map(round, f))
i
```

2. 사용자 정의 함수

```python
def num_max(a,b,c):
    max=a
    if max<b:
        max=b
    elif max<c:
        max=c
    return max

num = num_max(-3,-5,-1)
num
```
