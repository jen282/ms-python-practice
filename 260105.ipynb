```python
# 업앤다운 게임 만들어보기
import random
answer = random.randint(1,100)
cnt=0

while True:
    num = int(input('추측한 숫자를 입력해주세요'))
    if cnt > 10:
        print(f'10번 실패했어요! 게임 종료!')
    elif num < answer:
        print("업!")
        cnt+=1
    elif num > answer:
        print('다운!')
        cnt+=1
    elif num == answer:
        print(f'정답!! {cnt}번 만에 성공!')
        break
        
```

```python
# 미션
# 사용자로부터 n개의 별의 개수를 입력 받을 것
# 별을 출력하되 별 개수만큼 출력하다 별이 소진되면 멈추는 코드 작성
# 트리의 코드는 최대 10층

n = int(input('별의 개수를 입력해주세요'))

level, cnt=1, 0

while True:
    for i in range(1,level+1):
        if cnt==n:
            break
        print("*", end='')
        cnt+=1
        
    if cnt ==n or level ==10:
        break
    print()
    level +=1    
```

```python
n = int(input('별의 개수를 입력해주세요'))
for i in range (1, 11): #10층까지만 출력
    for i in range (i): #i층엔 i개 별까지 출력
        if n>0:
            print("*", end="")
            n -= 1 #출력할 때마다 개수 차감
    print()
    if n==0:
        break
```

```python
# 미션
marks = [90, 25, 67, 45, 80]

#아래와 같이 출력되도록 코드 작성 

for i in range(len(marks)):
    if marks[i] > 60:
        print(f'{i+1}번 학생은 합격입니다.')
    else:
        print(f'{i+1}번 학생은 불합격입니다.')
```

```python
# 미션
marks = [90, 25, 67, 45, 80]
num = 0
#아래와 같이 출력되도록 코드 작성 
for i in marks:
    num+=1
    if i>=60:
        print(f'{num}번 학생은 합격입니다.')
    else:
        print(f'{num}번 학생은 불합격입니다.')
```
